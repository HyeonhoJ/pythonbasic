```python
import xmltodict
import requests
sitemaps = rp.site_maps()

sitemaps


url = sitemaps[0]    #sitemap/latest-articles

sitemaps = xmltodict.parse(requests.get(url).text)

sitemaps['sitemapindex']['sitemap']

only_url = [sitemap['loc'] for sitemap in sitemaps['sitemapindex']['sitemap']] #sitemap url을 확보


only_url

response = requests.get(only_url[0])

urls = xmltodict.parse(response.text)

```

```python
news_urls = [urls['loc'] for urls in urls['urlset']['url']]
```

```python
%%time

# session 추상화
session = requests.Session()


for url in news_urls[:5]:
    file = url.split("/")[-2] + '.html'
    
    
    response = session.get(url)
    
    if response.ok:
        with open('datas/'+ file, 'w+b') as f:
            f.write(response.text.encode('utf-8'))
    else:
        print(f'error with URL : {url}')
```

```python
# os 패키지의 lisrdir() 메서드를 활용하면 arguments로 전달된 path 안의 파일 목록을 가져옴.

import os
path = './datas/'
files = [path + file for file in os.listdir(path)]
files
# HTML 파서를 사용한 데이터 추출
with open(files[1], 'r', encoding='utf-8') as f:
    html = f.read()
```
